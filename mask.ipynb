```python
import cv2
import time
import numpy as np
# Load the cascade
face_cascade = cv2.CascadeClassifier('harcas\haarcascade_frontalface_default.xml')

mouth_cascade = cv2.CascadeClassifier('haarcascade_mcs_mouth.xml')

 
cap = cv2.VideoCapture(0)


while True:
    # Read the frame
    _, img = cap.read()
    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Detect the faces
    faces = face_cascade.detectMultiScale(gray,1.03, 4)

    
    # Draw the rectangle around each face
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]

        mouths = mouth_cascade.detectMultiScale(roi_gray,1.1, 20)
#       print(mouths,type(mouths))
        for (ax, ay, aw, ah) in mouths:
            cv2.rectangle(roi_color, (ax, ay), (ax+aw, ay+ah), (0, 255,0 ), 3)
        if type(mouths)==tuple:
            print("mask")
        else:
            print("no-mask")
            
       
    # Display
    cv2.imshow('scanning', img)
    
    # Stop if escape key is pressed
    k = cv2.waitKey(30) & 0xff
    if k==27:
        break
# Release the VideoCapture object
cap.release()
cv2.destroyAllWindows
```
